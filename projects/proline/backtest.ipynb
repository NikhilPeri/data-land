```python
import os
os.chdir('/Users/nikhilperi/github.com/NikhilPeri/data-land')

import pandas as pd
from projects.proline.model import Predict, OUTPUT_COLUMNS
```

```python
odds = pd.read_csv('data/proline/odds.csv').drop_duplicates()
results = pd.read_csv('data/proline/results.csv').drop_duplicates()

results = results[
    (results['outcome_h_plus'] == 1) |
    (results['outcome_h'] == 1) |
    (results['outcome_t'] == 1) |
    (results['outcome_v'] == 1) |
    (results['outcome_v_plus'] == 1)
]
odds = odds[odds['sport'] != 'SCR']
odds = pd.merge(results, odds, on=['ticket_handle', 'game_handle'], how='inner').drop(OUTPUT_COLUMNS, axis=1).reset_index(drop=True)
```

```python
def backtest(tickets):
    actual_payout = []
    for _, ticket in tickets.iterrows():
        outcomes = results[(results['ticket_handle'] == ticket['ticket_handle']) & (results['game_handle'].isin(ticket['selected_games']))]
        win = True
        for index, selection in enumerate(ticket['selected_outcomes']):
            win &= (outcomes[outcomes['game_handle'] == ticket['selected_games'][index]][selection].values == [1]).all()
        actual_payout.append(ticket['payouts'].sum() if win else -1.0)

    predicted_tickets['actual_payout'] = actual_payout
    print "total profit: {}".format(predicted_tickets['actual_payout'].sum())
    display(predicted_tickets.head(100))
```

# Machine Learning

```python
predicted_outcomes = Predict().predict_outcomes(odds.copy())
predicted_tickets = Predict().compute_best_ticket_combinations(predicted_outcomes, max_games_to_consider=20)
None
```

```python
backtest(predicted_tickets)
```
