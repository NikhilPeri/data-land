```python
import os
import pandas as pd
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
from plotly.graph_objs import *

from dataland.storage import storage

PRICE_DATASET='data/statscanada/average-retail-prices-for-food-and-other-selected-items.csv'

os.chdir('/home/nikhil/github.com/NikhilPeri/data-land')
storage.pull(PRICE_DATASET)
price_data = pd.read_csv(storage.local_path(PRICE_DATASET))

init_notebook_mode(connected=True)
```

```python
price_data['com'].unique()
```

```python
def plot_commodities(commodities):
    comms = {}
    for com in commodities:
        data = price_data[price_data['com'] == com]
        comms[com] = Scatter(x=data['ref_date'], y=data['value'], name=com)
    iplot(comms.values())
```

```python
plot_commodities([
    'Paper towels (2 rolls)',
    'Facial tissue (200 tissues)', 
    'Bathroom tissue (4 rolls)',
    'Shampoo, 300 millilitres', 
    'Deodorant, 60 grams',
    'Toothpaste, 100 millilitres',
])
```
