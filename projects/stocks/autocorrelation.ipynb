```python
import os
os.chdir('/Users/nikhilperi/github.com/NikhilPeri/data-land')

import time
import pandas as pd
from projects.stocks.api import get_daily_prices

companies = pd.read_csv('data/stocks/snp_500_list.csv')
```

```python
RATE_LIMITER = 2

for index, company in companies.iloc[39:].iterrows():
    prices = get_daily_prices(company['Symbol'], full=True)
    time.sleep(RATE_LIMITER)
    if len(prices) < 5:
        print 'ERROR on {}'.format(index)
        time.sleep(10*RATE_LIMITER)
        prices = get_daily_prices(company['Symbol'], full=True)
            
    prices.to_csv('data/stocks/prices/{}.csv'.format(company['Symbol']), mode='w+', index=False)

    if index % 1 == 0:
        print "processed {}".format(index)
    
    if (index + 1) % 4 == 0:
        time.sleep(10*RATE_LIMITER)
        
```
