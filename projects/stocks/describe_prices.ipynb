# Intraday Stats

Our goal is to maximize the intraday mean while minimizing the intraday standard deviation. This will help us find a stock that trades with high daily change but with low day to day volitility, we can use the following scoring function:

```
score = weight_daily_change*normalized(intraday_mean) - weight_volitility*normalized(intraday_stddev)
```

````python
WEIGHT_DAILY_CHANGE = 1.0
WEIGHT_VOLITILITY = -1.0

stock_list['intraday_score'] = WEIGHT_DAILY_CHANGE * minmax_scale(stock_list['intraday_mean']) - WEIGHT_VOLITILITY * minmax_scale(stock_list['intraday_stddev'])
stock_list.sort_values('intraday_score', ascending=False).head(10)
````
